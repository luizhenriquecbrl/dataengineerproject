```python
import duckdb
```

```python
con = duckdb.connect(database='dados_duckdb.db',read_only=False)
```

```python
#select dos dados na camada silve
df = con.execute("SELECT * from produtos").fetchdf()
df.head(10)
```

```python
#criar dimensão de produtos na camada gold - remover colunas que não são da dimensão produtos
df_produtos = df.drop(columns=['ID_CATEGORIA','ID_FORNECEDOR'])
df_produtos.head(10)
```

```python
#criação da tabela dim_produtos
con.execute("""
    CREATE TABLE IF NOT EXISTS dim_produtos(
            ID_PRODUTO BIGINT,
            NM_PRODUTO TEXT,
            VL_PRECO FLOAT)
""")
```

```python
#inserção dos dados na dim_produtos
con.execute("INSERT INTO dim_produtos SELECT * FROM df_produtos")
```

```python
df_dim_produtos = con.execute("SELECT * FROM dim_produtos").fetchdf()
df_dim_produtos.head(10)
```
